```python
import numpy as np
import pandas as pd
import seaborn as sns
from lifelines import WeibullFitter
```

```python
np.random.seed(0)
sns.set_theme()
```

```python
df = pd.read_csv(
    "../data/kickstarter_1.csv",
    parse_dates=True,
)
```

```python
event_times = df["day_succ"]
event_observed = df["Status"]
```

```python
wbf = WeibullFitter()
wbf.fit(event_times, event_observed)
wbf.plot()
wbf.plot_survival_function()
```
